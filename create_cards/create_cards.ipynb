```python
import os
import numpy as np
from numpy import pi as pi
from PIL import Image, ImageDraw, ImageFont, ImageOps

filenames = ['bass', 'diamond', 'spades', 'heart', 'clubs']
jokers = ['scooter', 'HBz', 'The Holy Santa Barbara', 'Rammstein', 'paling', 'BMTH']
cards = ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'B', 'D', 'K', 'A']

output = 'cards2/'
inverse = False


font_str = 'BOD_CB.TTF'
font = ImageFont.truetype(font_str, 70)
font_big = ImageFont.truetype(font_str, 200)


input = 'resources/inverse/' if inverse else 'resources/'
background = 'black' if inverse  else 'white'
foreground = 'black' if not inverse  else 'white'


# Card
card_h = 1146 #900
card_w = 769 #600

joker_img_w = int(350 / 600 * card_w)

h_max_smol = 0
h_max_big = 0

####### OUTSIDE #######
out_w = int(55 / 600 * card_w)
out_h_txt = 70
out_margin = 50

if 'clubs' in filenames: filenames.append('clubs_rgb')
for filename in filenames:
    img = Image.open(input + filename + '.png')
    wpercent = (out_w/float(img.size[0]))
    hsize = int((float(img.size[1])*float(wpercent)))
    img = img.resize((out_w,hsize), Image.Resampling.LANCZOS)
    img.save(input + filename + '_smol.png')
    if hsize > h_max_smol:
        h_max_smol = hsize

def AdjustYPos_smol(img, pos, flip = False) -> int:
    if flip:
        return (pos[0], pos[1] - int((h_max_smol - img.height) / 2))
    else:
        return (pos[0], pos[1] + int((h_max_smol - img.height) / 2))


####### INSIDE #######
in_w = int(100 / 600 * card_w)
crown_w = 60

for filename in filenames:
    img = Image.open(input + filename + '.png')
    wpercent = (in_w/float(img.size[0]))
    hsize = int((float(img.size[1])*float(wpercent)))
    img = img.resize((in_w,hsize), Image.Resampling.LANCZOS)
    img.save(input + filename + '_big.png')
    if hsize > h_max_smol:
        h_max_big = hsize

def AdjustYPos_big(img, pos, flip = False) -> int:
    if flip:
        return (pos[0], pos[1] - int((h_max_big - img.height) / 2))
    else:
        return (pos[0], pos[1] + int((h_max_big - img.height) / 2))


if 'clubs_rgb' in filenames: filenames.remove('clubs_rgb')

# CROWN 
for filename in filenames:
    img = Image.open(input + 'king_' + filename +'.png')
    wpercent = (crown_w/float(img.size[0]))
    hsize = int((float(img.size[1])*float(wpercent)))
    img = img.resize((crown_w,hsize), Image.Resampling.LANCZOS)
    img.save(input + 'king_' + filename +'_smol.png')

####### ACE #######
ace_img_w = int(200 / 600 * card_w)
for filename in filenames:
    # if filename == 'bass':
    #     img = Image.open(input + filename + '_big_boosted2.png')
    # else:
    img = Image.open(input + filename + '.png')
    wpercent = (ace_img_w/float(img.size[0]))
    hsize = int((float(img.size[1])*float(wpercent)))
    img = img.resize((ace_img_w,hsize), Image.Resampling.LANCZOS)
    img.save(input + filename + '_ace.png')

# BACK
back_w = 180
for filename in filenames:
    img = Image.open(input + filename +'_mit_rand.png')
    wpercent = (back_w/float(img.size[0]))
    hsize = int((float(img.size[1])*float(wpercent)))
    img = img.resize((back_w,hsize), Image.Resampling.LANCZOS)
    img.save(input + filename +'_mit_rand_sized.png')



if not os.path.exists(output):
    os.mkdir(output)
```

```python
####### ALL REGULAR CARDS #######

for color in filenames:

    color_out = Image.open(input + color +'_smol.png')
    color_in = Image.open(input + color +'_big.png')
    
    sym_tl = (out_margin, out_margin + out_h_txt)
    sym_tr = (card_w - out_margin - color_out.width, out_margin + out_h_txt)
    sym_bl = (out_margin, card_h - out_margin - out_h_txt - color_out.height)
    sym_br = (card_w - out_margin - color_out.width, card_h - out_margin - out_h_txt - color_out.height)

    # Calculate positions
    card_w_cen =int(card_w/2 - color_in.width/2)
    card_h_cen = int(card_h/2 - color_in.height/2)

    in_margin_h = int(card_h/2 - 200)
    in_margin_w = int(card_w/2 - 230)


    pos_1 = (card_w_cen,                int(card_h_cen - 1 * in_margin_h))
    pos_2 = (card_w_cen,                int(card_h_cen + 1 * in_margin_h))
    pos_3 = (card_w_cen,                card_h_cen)
    pos_4 = (card_w_cen - in_margin_w,  int(card_h_cen - 1 * in_margin_h))
    pos_5 = (card_w_cen + in_margin_w,  int(card_h_cen - 1 * in_margin_h))
    pos_6 = (card_w_cen - in_margin_w,  int(card_h_cen + 1 * in_margin_h))
    pos_7 = (card_w_cen + in_margin_w,  int(card_h_cen + 1 * in_margin_h))
    pos_8 = (card_w_cen - in_margin_w,  card_h_cen)
    pos_9 = (card_w_cen + in_margin_w,  card_h_cen)
    pos_10 = (card_w_cen,               int(card_h_cen - 1/2 * in_margin_h))
    pos_11 = (card_w_cen,               int(card_h_cen + 1/2 * in_margin_h))
    pos_12 = (card_w_cen - in_margin_w, int(card_h_cen - 1/3 * in_margin_h))
    pos_13 = (card_w_cen + in_margin_w, int(card_h_cen - 1/3 * in_margin_h))
    pos_14 = (card_w_cen - in_margin_w, int(card_h_cen + 1/3 * in_margin_h))
    pos_15 = (card_w_cen + in_margin_w, int(card_h_cen + 1/3 * in_margin_h))
    pos_16 = (card_w_cen,               int(card_h_cen - 2/3 * in_margin_h))
    pos_17 = (card_w_cen,               int(card_h_cen + 2/3 * in_margin_h))

    positions = [
        [pos_1, pos_2], # 2
        [pos_1, pos_2, pos_3], #3
        [pos_4, pos_5, pos_6, pos_7], # 4
        [pos_3, pos_4, pos_5, pos_6, pos_7], # 5
        [pos_4, pos_5, pos_6, pos_7, pos_8, pos_9], # 6
        [pos_4, pos_5, pos_6, pos_7, pos_8, pos_9, pos_10], # 7
        [pos_4, pos_5, pos_6, pos_7, pos_8, pos_9, pos_10, pos_11], # 8
        [pos_3, pos_4, pos_5, pos_6, pos_7, pos_12, pos_13, pos_14, pos_15], # 9
        [pos_4, pos_5, pos_6, pos_7, pos_12, pos_13, pos_14, pos_15, pos_16, pos_17], # 10
    ]

    font_positions = [
        (out_margin, out_margin),
        (card_w - out_margin - out_w, out_margin),
        (out_margin, card_h - out_margin - out_h_txt),
        (card_w - out_margin - out_w, card_h - out_margin  - out_h_txt)
    ]

    for number in cards:

        if color == 'clubs' and number == '7':
            temp_color_in = color_in
            temp_color_out = color_out
            color_out = Image.open(input + 'clubs_rgb_smol.png')
            color_in = Image.open(input + 'clubs_rgb_big.png')

        # Create a blank canvas for the card
        card = Image.new('RGB', (card_w, card_h), color=background)

        # Outside Symbols
        card.paste(color_out, AdjustYPos_smol(color_out, sym_tl), mask=color_out.split()[3])
        card.paste(color_out.transpose(Image.FLIP_LEFT_RIGHT), AdjustYPos_smol(color_out, sym_tr), mask=color_out.transpose(Image.FLIP_LEFT_RIGHT).split()[3])
        card.paste(color_out.rotate(180).transpose(Image.FLIP_LEFT_RIGHT), AdjustYPos_smol(color_out, sym_bl, True), mask=color_out.rotate(180).transpose(Image.FLIP_LEFT_RIGHT).split()[3])
        card.paste(color_out.rotate(180), AdjustYPos_smol(color_out, sym_br, True), mask=color_out.rotate(180).split()[3])        

        # Outside Text
        for pos in font_positions:
            txt=Image.new('RGB', (out_w, out_h_txt), color=background)
            txt_w = font.getlength(number)
            d = ImageDraw.Draw(txt)
            d.text((int(out_w / 2 - txt_w / 2), 0), number,  font=font, fill=foreground)
            if pos[1] > card_h_cen:
                w = txt.rotate(180)
            else:
                w = txt
            card.paste(w, pos)

        # Inside Symbols
        if number.isnumeric():
            for pos in positions[int(number) - 2]:
                if pos[1] > card_h_cen:
                    card.paste(color_in.rotate(180), AdjustYPos_big(color_in, pos, True), mask=color_in.rotate(180).split()[3])
                else:
                    card.paste(color_in, AdjustYPos_big(color_in, pos), mask=color_in.split()[3])
        
        elif number == 'A':
            color_ace = Image.open(input + color + '_ace.png')
            card.paste(color_ace, (int(card_w/2 - color_ace.width/2), int(card_h/2 - color_ace.height/2)), mask=color_ace.split()[3])
        
        elif number == 'K':
            joker_img = Image.open(input + 'paling_r_joker.png')
            card.paste(joker_img, (int(card_w/2 - joker_img.width/2), int(card_h/2 - joker_img.height/2)), mask=joker_img.split()[3])
            crown_img = Image.open(input + 'king_' + color + '_smol.png')
            card.paste(crown_img, (180, 530), mask=crown_img.split()[3])
        
        elif number == 'D':
            joker_img = Image.open(input + 'paling_g_joker.png')
            card.paste(joker_img, (int(card_w/2 - joker_img.width/2), int(card_h/2 - joker_img.height/2)), mask=joker_img.split()[3])
        
        elif number == 'B':
            joker_img = Image.open(input + 'paling_b_joker.png')
            card.paste(joker_img, (int(card_w/2 - joker_img.width/2), int(card_h/2 - joker_img.height/2)), mask=joker_img.split()[3])

        
        if color == 'clubs' and number == '7':
            color_in = temp_color_in
            color_out = temp_color_out

        # Save the final image as a file
        card.save(output + color + '_' + number + '.png')
```

```python
####### JOKERS #######
for idx, joker in enumerate(jokers):

    joker_img = Image.open(input + joker + '.png')
    wpercent = (joker_img_w/float(joker_img.size[0]))
    hsize = int((float(joker_img.size[1])*float(wpercent)))
    joker_img.resize((joker_img_w,hsize), Image.Resampling.LANCZOS).save(input + joker + '_joker.png')
    joker_img = Image.open(input + joker + '_joker.png').convert('RGBA')

    text = 'J\nO\nK\nE\nR'
    card = Image.new('RGB', (card_w, card_h), color=background)
    out_w_joker = font.getbbox(text)[3]
    out_h_joker = font.getbbox(text)[2]

    # Outside text
    for pos in [(out_margin + 15, out_margin), (card_w - out_margin - out_w_joker, card_h - out_margin  - out_h_joker)]:
        txt=Image.new('RGB', (out_w, out_h_joker), color=background)
        d = ImageDraw.Draw(txt)
        d.text((int(out_w / 2 - txt_w / 2), 0), text,  font=font, fill=foreground)
        if pos[1] > card_h_cen:
            w = txt.rotate(180)
        else:
            w = txt
        card.paste(w, pos)

    # Inside symbol
    card.paste(joker_img, (int(card_w/2 - joker_img.width/2), int(card_h/2 - joker_img.height/2)), mask=joker_img.split()[3])

    card.save(output + 'joker_' + joker + '.png')
```

```python
####### BACK #######
back_spam = 'LTT_sw'
back_spam_w = 40


card = Image.new('RGB', (card_w, card_h), color=(70, 70, 70))

# Spam symbol
back_sym = Image.open(input + back_spam + '.png')
wpercent = (back_spam_w/float(back_sym.size[0]))
hsize = int((float(back_sym.size[1])*float(wpercent)))
back_sym = back_sym.resize((back_spam_w,hsize), Image.Resampling.LANCZOS)

cols = int(card_w / back_spam_w)
rows = int(card_h / hsize)
back_offset_w = int(int(card_w - cols * back_spam_w) / 2)
back_offset_h = int(int(card_h - rows * hsize) / 2)

for x in range(cols + 2):
    for y in range(rows + 2):
        card.paste(back_sym, (x * back_spam_w - back_offset_w, y * hsize - back_offset_h), mask=back_sym.split()[3])


# Main symbols
back_margin = card_w / 3
for idx, symbol in enumerate(filenames):
    angle = idx / len(filenames) * 360
    x = int(card_w_cen - back_margin * np.sin(angle * pi / 180))
    y = int(card_h_cen - back_margin * np.cos(angle * pi / 180))
    current_pos = (x, y)
    
    symbol_img = Image.open(input + symbol +'_mit_rand_sized.png')
    card.paste(symbol_img, AdjustYPos_big(symbol_img, current_pos), mask=symbol_img.split()[3])

card.save(output + 'back.png')
```

```python
####### FULL DECK #######

h = 0
w = 0
margin = 100

canvas = Image.new('RGB', (max(len(cards), len(jokers)) * (card_w + margin) - margin, (len(filenames) + int(len(jokers) > 0)) * (card_h + margin) - margin), color=foreground)

for color in filenames:
    for number in cards:
        card = Image.open(output + color + '_' + number + '.png')
        canvas.paste(card, (w, h))
        w += card_w + margin
    
    h += card_h + margin
    w = 0

for joker in jokers:
    card = Image.open(output + 'joker_' + joker + '.png')
    canvas.paste(card, (w, h))
    w += card_w + margin

canvas.paste(Image.open(output + 'back.png'), (w, h))

canvas.save(output + 'canvas.png')
```

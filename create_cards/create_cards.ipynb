```python
from PIL import Image, ImageDraw, ImageFont, ImageOps

filenames = ['clubs', 'spades', 'heart', 'diamond', 'bass']
cards = ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'B', 'D', 'K', 'A']
output = 'cards2/'
input = 'resources/inverse/'
background = 'black'
foreground = 'white'
jokers = 5

font_str = 'BOD_CB.TTF'
font = ImageFont.truetype(font_str, 70)
font_big = ImageFont.truetype(font_str, 200)

# Card
card_h = 900
card_w = 600

####### OUTSIDE #######
out_w = 55
out_h_txt = 70
out_margin = 20

if 'clubs' in filenames: filenames.append('clubs_rgb')
for filename in filenames:
    img = Image.open(input + filename + '.png')
    wpercent = (out_w/float(img.size[0]))
    hsize = int((float(img.size[1])*float(wpercent)))
    img = img.resize((out_w,hsize), Image.Resampling.LANCZOS)
    img.save(input + filename + '_smol.png')


####### INSIDE #######
in_margin_h = 160
in_margin_w = 130
in_w = 120

for filename in filenames:
    img = Image.open(input + filename + '.png')
    wpercent = (in_w/float(img.size[0]))
    hsize = int((float(img.size[1])*float(wpercent)))
    img = img.resize((in_w,hsize), Image.Resampling.LANCZOS)
    img.save(input + filename + '_big.png')

if 'clubs_rgb' in filenames: filenames.remove('clubs_rgb')
```

```python
####### ALL REGULAR CARDS #######

for color in filenames:

    color_out = Image.open(input + color +'_smol.png')
    color_in = Image.open(input + color +'_big.png')
    
    sym_tl = (out_margin, out_margin + out_h_txt)
    sym_tr = (card_w - out_margin - color_out.width, out_margin + out_h_txt)
    sym_bl = (out_margin, card_h - out_margin - out_h_txt - color_out.height)
    sym_br = (card_w - out_margin - color_out.width, card_h - out_margin - out_h_txt - color_out.height)

    # Calculate positions
    card_w_cen =int(card_w/2 - color_in.width/2)
    card_h_cen = int(card_h/2 - color_in.height/2)

    pos_1 = (card_w_cen, card_h_cen - 2 * in_margin_h)
    pos_2 = (card_w_cen, card_h_cen + 2 * in_margin_h)
    pos_3 = (card_w_cen, card_h_cen)
    pos_4 = (card_w_cen - in_margin_w, card_h_cen - 2 * in_margin_h)
    pos_5 = (card_w_cen + in_margin_w, card_h_cen - 2 * in_margin_h)
    pos_6 = (card_w_cen - in_margin_w, card_h_cen + 2 * in_margin_h)
    pos_7 = (card_w_cen + in_margin_w, card_h_cen + 2 * in_margin_h)
    pos_8 = (card_w_cen - in_margin_w, card_h_cen)
    pos_9 = (card_w_cen + in_margin_w, card_h_cen)
    pos_10 = (card_w_cen, card_h_cen - in_margin_h)
    pos_11 = (card_w_cen, card_h_cen + in_margin_h)
    pos_12 = (card_w_cen - in_margin_w, int(card_h_cen - 0.75 * in_margin_h))
    pos_13 = (card_w_cen + in_margin_w, int(card_h_cen - 0.75 * in_margin_h))
    pos_14 = (card_w_cen - in_margin_w, int(card_h_cen + 0.75 * in_margin_h))
    pos_15 = (card_w_cen + in_margin_w, int(card_h_cen + 0.75 * in_margin_h))
    pos_16 = (card_w_cen, int(card_h_cen - 1.4 * in_margin_h))
    pos_17 = (card_w_cen, int(card_h_cen + 1.4 * in_margin_h))

    positions = [
        [pos_1, pos_2], # 2
        [pos_1, pos_2, pos_3], #3
        [pos_4, pos_5, pos_6, pos_7], # 4
        [pos_3, pos_4, pos_5, pos_6, pos_7], # 5
        [pos_4, pos_5, pos_6, pos_7, pos_8, pos_9], # 6
        [pos_4, pos_5, pos_6, pos_7, pos_8, pos_9, pos_10], # 7
        [pos_4, pos_5, pos_6, pos_7, pos_8, pos_9, pos_10, pos_11], # 8
        [pos_3, pos_4, pos_5, pos_6, pos_7, pos_12, pos_13, pos_14, pos_15], # 9
        [pos_4, pos_5, pos_6, pos_7, pos_12, pos_13, pos_14, pos_15, pos_16, pos_17], # 10
    ]

    font_positions = [
        (out_margin, out_margin),
        (card_w - out_margin - out_w, out_margin),
        (out_margin, card_h - out_margin - out_h_txt),
        (card_w - out_margin - out_w, card_h - out_margin  - out_h_txt)
    ]

    for number in cards:

        if color == 'clubs' and number == '7':
            temp_color_in = color_in
            temp_color_out = color_out
            color_out = Image.open('resources/clubs_rgb_smol.png')
            color_in = Image.open('resources/clubs_rgb_big.png')

        # Create a blank canvas for the card
        card = Image.new('RGB', (card_w, card_h), color='white')

        # Outside Symbols
        card.paste(color_out, sym_tl, mask=color_out.split()[3])
        card.paste(color_out.transpose(Image.FLIP_LEFT_RIGHT), sym_tr, mask=color_out.transpose(Image.FLIP_LEFT_RIGHT).split()[3])
        card.paste(color_out.rotate(180).transpose(Image.FLIP_LEFT_RIGHT), sym_bl, mask=color_out.rotate(180).transpose(Image.FLIP_LEFT_RIGHT).split()[3])
        card.paste(color_out.rotate(180), sym_br, mask=color_out.rotate(180).split()[3])        

        # Outside Text
        for pos in font_positions:
            txt=Image.new('RGB', (out_w, out_h_txt), color='white')
            txt_w = font.getlength(number)
            d = ImageDraw.Draw(txt)
            d.text((int(out_w / 2 - txt_w / 2), 0), number,  font=font, fill='black')
            if pos[1] > card_h_cen:
                w = txt.rotate(180)
            else:
                w = txt
            card.paste(w, pos)

        # Inside Symbols
        if number.isnumeric():
            for pos in positions[int(number) - 2]:
                if pos[1] > card_h_cen:
                    card.paste(color_in.rotate(180), pos, mask=color_in.rotate(180).split()[3])
                else:
                    card.paste(color_in, pos, mask=color_in.split()[3])
        elif number == 'A':
            card.paste(color_in, (card_w_cen, card_h_cen), mask=color_in.split()[3])
        
        if color == 'clubs' and number == '7':
            color_in = temp_color_in
            color_out = temp_color_out

        # Save the final image as a file
        card.save(output + color + '_' + number + '.png')
```

```python
####### JOKERS #######
for index in range(jokers):

    joker_img_str = 'scooter'
    joker_img = Image.open(input + joker_img_str + '.png')
    joker_img_w = 400
    wpercent = (joker_img_w/float(joker_img.size[0]))
    hsize = int((float(joker_img.size[1])*float(wpercent)))
    joker_img.resize((joker_img_w,hsize), Image.Resampling.LANCZOS).save(input + joker_img_str + '_joker.png')
    joker_img = Image.open(input + joker_img_str + '_joker.png').convert('RGBA')

    text = 'J\nO\nK\nE\nR'
    card = Image.new('RGB', (card_w, card_h), color='white')
    out_w_joker = font.getbbox(text)[3]
    out_h_joker = font.getbbox(text)[2]

    # Outside text
    for pos in [(out_margin + 15, out_margin), (card_w - out_margin - out_w_joker - 15, card_h - out_margin  - out_h_joker)]:
        txt=Image.new('RGB', (out_w, out_h_joker), color='white')
        d = ImageDraw.Draw(txt)
        d.text((int(out_w / 2 - txt_w / 2), 0), text,  font=font, fill='black')
        if pos[1] > card_h_cen:
            w = txt.rotate(180)
        else:
            w = txt
        card.paste(w, pos)

    # Inside symbol
    card.paste(joker_img, (int(card_w/2 - joker_img.width/2), int(card_h/2 - joker_img.height/2)), mask=joker_img.split()[3])

    card.save(output + 'joker_' + str(index + 1) + '.png')
```
